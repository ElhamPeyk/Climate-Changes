```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path=("climate_change_indicators.csv")
data=pd.read_csv("climate_change_indicators.csv")
```

```python
country_name = "Germany"
country_data = data[data["Country"] == country_name]
```

```python
year_columns = [col for col in country_data.columns if col.startswith("F")]
```

```python
years = [int(col[1:]) for col in year_columns]
```

```python
temperature_changes = country_data[year_columns].iloc[0]
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.plot(years, temperature_changes, marker='o', label=country_name)
plt.title(f"Temperature Changes Over Time in {country_name}", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Temperature Change (Â°C)", fontsize=12)
plt.grid(alpha=0.5)
plt.legend()
plt.show()

```

```python
countries = ["Germany", "France", "Italy", "Spain"]
```

```python
comparison_data = data[data["Country"].isin(countries)]
```

```python
plt.figure(figsize=(14, 7))
for country in countries:
    country_data = comparison_data[comparison_data["Country"] == country]
    temperature_changes = country_data[year_columns].iloc[0]
    plt.plot(years, temperature_changes, marker='o', label=country)
    plt.title("Temperature Changes Over Time in Different Countries", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Temperature Change (Â°C)", fontsize=14)
plt.legend(title="Country")
plt.grid(alpha=0.5)
plt.show()
```
